# 1.Інкапсуляція
### До прикладу з роботи додав генератор випадкових чисел та викликав методи deposit та withdraw у циклі передаючи випадкові числа. А також викликав кінцевий результат; 

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  
        self.__balance = balance  

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print(f"Attempted to withdraw {amount} - Insufficient funds")

    def get_balance(self):
        return self.__balance

account = BankAccount("Maksym", 1000)
for i in range(10):
    amount = random.randint(1, 500)  
    
    if random.choice([True, False]):
        account.deposit(amount)
    else:
        account.withdraw(amount)

print("\nFinal balance:", account.get_balance())
```

# 2.Наслідування
### Створив ще один метод у класі Vehicle та викликав його з обєкта класу Car;

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_engine(self):
        return f"{self.brand} {self.model} engine started!"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

car = Car("Toyota", "Camry", 5)

print(car.display_info())
print(car.start_engine())    
```

# 3. Поліморфизм
### Створив клас Fish у якого немає методу speak, як результат виводить None

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Fish(Animal):
    pass   # немає власного методу speak()


animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(animal.speak())
```

# 4. Абстракція
### В даному прикладі не було завдання тому переписав код з роботи

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())
```

# 5. Використання парадигм для створення простої гри
### було додано третій тип зброї Bow із власним параметром дальності та методом перезарядки, реалізовано заточку для сокири, запроваджено випадковий вибір зброї для гравця і бота, створено покрокову інтерактивну систему бою з вибором дій користувачем, додано перевірку стану здоров’я для завершення гри та вдосконалено бойову механіку із застосуванням принципів абстракції, наслідування, інкапсуляції та поліморфізму.

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health 

    @abstractmethod
    def attack(self, another_item):
        pass

    def is_alive(self):
        return self.health > 0




class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name)
        self.__attack_power = attack_power
        self._sharp = 0

    def sharpening(self):
        self._sharp += 10
        return f"{self.name} загострено. Sharpness: {self._sharp}"

    def attack(self, another_item):
        damage = self.__attack_power + self._sharp + randint(0, 10) 
        another_item.health = max(0, another_item.health - damage)
        return f"Меч {self.name} наносить {damage} шкоди. HP суперника: {another_item.health}"



class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name)
        self.__attack_power = attack_power
        self._sharp = 0  

    def sharpening_axe(self):
        self._sharp += 20  
        return f"{self.name} заточено. Sharpness: {self._sharp}"

    def attack(self, another_item):
        damage = self.__attack_power + self._sharp + randint(0, 20) 
        another_item.health = max(0, another_item.health - damage)
        return f"Сокира {self.name} наносить {damage} шкоди. HP суперника: {another_item.health}"

    @property
    def get_attack_power(self):
        return f"Атака сокири {self.name}: {self.__attack_power + self._sharp}"


class Bow(Item):
    def __init__(self, name, attack_power: int, range_power: int):
        super().__init__(name)
        self.__attack_power = attack_power
        self.range_power = range_power

    def reload(self):
        self.range_power += 1
        return f"{self.name} підвищує дальність. Range: {self.range_power}"

    def attack(self, another_item):
        damage = self.__attack_power + randint(5, 15) + self.range_power 
        another_item.health = max(0, another_item.health - damage)
        return f"Лук {self.name} наносить {damage} шкоди. HP суперника: {another_item.health}"



def random_weapon(player_name):
    weapons = [
        Sword("Ескалібур", 100),
        Axe("Кратос", 95),
        Bow("Ельфійський лук", 80, 5)
    ]
    weapon = choice(weapons)
    weapon.name = f"{player_name} - {weapon.name}"
    return weapon


#  Гра
player = random_weapon("Гравець")
enemy = random_weapon("Бот")

print(f"Твоя зброя: {type(player).__name__}")
print(f"Зброя бота: {type(enemy).__name__}")

turn = 1

while player.is_alive() and enemy.is_alive():
    print(f"\n--- Хід {turn} ---")
    print("1 - Атакувати")

    if isinstance(player, Sword):
        print("2 - Загострити меч")
    elif isinstance(player, Bow):
        print("2 - Перезарядити лук")
    elif isinstance(player, Axe):
        print("2 - Заточити сокиру")

    choice_action = input("Обери дію: ")

    # ХІД ГРАВЦЯ
    if choice_action == "1":
        print(player.attack(enemy))
    elif choice_action == "2":
        if isinstance(player, Sword):
            print(player.sharpening())
        elif isinstance(player, Bow):
            print(player.reload())
        else:
            print(player.sharpening_axe())

    else:
        print("Невірний вибір!")

    if not enemy.is_alive():
        print(" ТИ ПЕРЕМІГ!")
        break

    # ХІД БОТА (випадкова дія)
    bot_action = randint(1, 2)

    if bot_action == 1:
        print(enemy.attack(player))
    else:
        if isinstance(enemy, Sword):
            print(enemy.sharpening())
        if isinstance(enemy, Axe):
            print(enemy.sharpening_axe())
        if isinstance(enemy, Bow):
            print(enemy.reload())
    if not player.is_alive():
        print(" БОТ ПЕРЕМІГ!")
        break

    turn += 1
```
